```python
import torch
from sklearn.datasets import make_regression

import matplotlib.pyplot as plt

# Generate simple linear regression data
X, y = make_regression(n_samples=100, n_features=1, noise=0.1)
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.float32)

# Plot the data
plt.scatter(X, y)
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Simple Linear Regression Data')
plt.show()
```

```python
from ActNet import ActNet
from utils import train
import torch.nn as nn
from torch.utils.data import DataLoader

d_in = 1
d_out = 1
d = 10
m = 5
N = 3
L = 2

model = ActNet(d_in, d_out, d, m, N, L)
loss_fn = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=1e-3)
data = DataLoader(list(zip(X, y)), batch_size=10, shuffle=True)

train(data, data, model, optimizer, loss_fn, n_epochs=100)
```
